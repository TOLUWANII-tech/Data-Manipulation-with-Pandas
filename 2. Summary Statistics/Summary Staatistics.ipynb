```python
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

sales[['temperature_c','fuel_price_usd_per_l', 'unemployment']].agg([iqr, np.median])
#The .agg() method allows you to carry out summary functions on a dataset or subsets of a dataset.
# It is particularly useful when you want to use a custom function that's not in the pandas or numpy library.
```

### Grouped Summary Statistics

```python
# Calculating the total weekly sales
sales_all = sales["weekly_sales"].sum()

# Subset for type A stores, calc total weekly sales
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()

# Subset for type B stores, calc total weekly sales
sales_B = sales[sales['type']=='B']['weekly_sales'].sum()

# Subset for type C stores, calc total weekly sales
sales_C = sales[sales['type']=='C']['weekly_sales'].sum()

# Get proportion for each type
sales_propn_by_type = [sales_A, sales_B, sales_C] /sales_all
print(sales_propn_by_type)
```

```python
#USING .groupby() method

sales_by_type = sales.groupby("type")["weekly_sales"].sum()
sales_by_type
```

```python
sales_by_type_proportion = sales_by_type/sum(sales['weekly_sales'])
sales_by_type_proportion
```

```python
#Grouping using values from multiple columns
sales_by_type_is_holiday = sales.groupby(['type', 'is_holiday'])['weekly_sales'].sum()
sales_by_type_is_holiday
```

```python
#Grouping and aggregating several functions on a column
sales_stats = sales.groupby('type')['weekly_sales'].agg([min, max, np.mean, np.median])
sales_stats
```

```python
unemployment_sales_stats = sales.groupby('type')[['unemployment','fuel_price_usd_per_l']].agg([min, max, np.mean, np.median])
unemployment_sales_stats
```

### PIVOT TABLES 
#### Altrenative method for summary statistics

```python
#Pivoting values on one variable.
#Claculating the mean of 'weekly_sales' grouped by 'type'
mean_sales_by_type = sales.pivot_table(values='weekly_sales', index='type')
mean_sales_by_type


#By default the .pivot_table() method finds the mean of the 'value' keyword argument. 
#To carry out other statistics on the 'value' argument, we pass in another argument to the methos, 'aggfunc' which accepts a list item of other statistical functions.
```

```python
#Getting the mean and median values of the 'weekly_sales' column grouped by 'type' using the 'aggfunc '
mean_med_sales_by_type = sales.pivot_table(values='weekly_sales', index='type', aggfunc=[np.mean, np.median])
mean_med_sales_by_type
```

```python
#Getting statistics of 'weekly_sales' by 'type' and 'is_holiday' using the 'columns' argument on .pivot_tableb

mean_med_sales_by_type_holiday = sales.pivot_table(values='weekly_sales', index='type', columns='is_holiday', aggfunc=[np.mean, np.median])
mean_med_sales_by_type_holiday
```

```python
#Filling missing values in te summary statistics using 'fill_value' argument
#Using the 'margins' argument to sum the statistics.
sales.pivot_table(values='weekly_sales', index='type', columns='department', fill_value = 0, margins = True)
```
